## Importing Library

```python
import pandas as pd

from sqlalchemy import create_engine
```

```python
FilmLocations=pd.read_csv('Film_Locations_in_San_Francisco.csv')
```

```python
FilmLocations.head(3)
```

```python
FilmLocations.info()
```

## Renaming Culumn without space

```python
FilmLocations=FilmLocations.rename(columns={"Release Year": "ReleaseYear","Fun Facts":"FunFacts", "Production Company": "ProductionCompany","Actor 1":"Actor1","Actor 2":"Actor2","Actor 3":"Actor3"})
```

```python
FilmLocations.info()
```

## Making SQL connection (from CSV)

```python
temp_db = create_engine('sqlite:///:memory:')
```

```python
FilmLocations.to_sql(name='FilmLocations',con=temp_db)
```

```python
pd.read_sql(sql='FilmLocations',con=temp_db)
```

# Exercise 2: DISTINCT

## Task A: Example exercises of DISTINCT

### Retrieve the name of all films without any repeated titles.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Title FROM FilmLocations
                        """,con=temp_db)
```

### Retrieve the number of release years of the films distinctly, produced by Warner Bros. Pictures.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT ReleaseYear FROM FilmLocations
                          WHERE ProductionCompany="Warner Bros. Pictures"
                        """,con=temp_db)
```

## Task B: Practice exercises on DISTINCT

### Retrieve the name of all unique films released in the 21st century and onwards, along with their release years.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Title, ReleaseYear FROM FilmLocations
                          WHERE ReleaseYear>2000
                        """,con=temp_db)
```

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT ReleaseYear, Title FROM FilmLocations
                          WHERE ReleaseYear>2000
                        """,con=temp_db)
```

### Retrieve the names of all the directors and their distinct films shot at City Hall.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Director, Title
                          FROM FilmLocations
                          WHERE Locations='City Hall'
                        """,con=temp_db)
```

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Title, Director
                          FROM FilmLocations
                          WHERE Locations='City Hall'
                        """,con=temp_db)
```

### Retrieve the number of distributors distinctly who distributed films acted by Clint Eastwood as 1st actor.

```python
pd.read_sql_query(sql="""
                        SELECT COUNT(DISTINCT Distributor) FROM FilmLocations
                        WHERE Actor1="Clint Eastwood"
                        """,con=temp_db)
```

# Exercise 3: LIMIT  and OFFSET

## Task A: Example exercises of LIMIT

### Retrieve the first 25 rows from the "FilmLocations" table.

```python
pd.read_sql_query(sql="""
                          SELECT * FROM FilmLOcations
                          LIMIT 25
                        """,con=temp_db)
```

### Retrieve the first 15 rows from the "FilmLocations" table starting from row 11.

```python
pd.read_sql_query(sql="""
                          SELECT * FROM FilmLOcations
                          LIMIT 15 OFFSET 10
                        """,con=temp_db)
```

## Task B: Practice exercises on LIMIT

### Retrieve the name of first 50 films distinctly.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Title FROM FilmLocations
                          LIMIT 50
                        """,con=temp_db)
```

### Retrieve first 10 film names distinctly released in 2015.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Title FROM FilmLOcations
                          WHERE ReleaseYear=2015
                          LIMIT 10
                        """,con=temp_db)
```

### Retrieve the next 3 film names distinctly after first 5 films released in 2015.

```python
pd.read_sql_query(sql="""
                          SELECT DISTINCT Title FROM FilmLocations
                          WHERE ReleaseYear=2015
                          LIMIT 3 OFFSET 5
                        """,con=temp_db)
```
