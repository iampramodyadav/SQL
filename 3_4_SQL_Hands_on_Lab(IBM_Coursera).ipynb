## adding first table

```python
#credit @geekforgeeks

# Import required libraries
import sqlite3
import pandas as pd

# Connect to SQLite database
conn = sqlite3.connect(r'dataset.db')

# Load CSV data into Pandas DataFrame
employ_data = pd.read_csv('dataset/emplyees.csv')
# Write the data to a sqlite table
employ_data.to_sql('EMPLOYEE', conn, if_exists='replace', index=False)

# Create a cursor object
cur = conn.cursor()
# Fetch and display result
for row in cur.execute('SELECT * FROM EMPLOYEE'):
	print(row)
# Close connection to SQLite database
conn.close()
```

# Printing schema 

```python
con = sqlite3.connect('dataset.db') 
pd.read_sql_query("SELECT * FROM sqlite_master where type = 'table';", con)
```

```python
tables = pd.read_sql_query("SELECT NAME AS 'Table_Name' FROM sqlite_master WHERE type='table'",con)
tables = tables["Table_Name"].values.tolist()
```

```python
for table in tables:
    query = "PRAGMA TABLE_INFO({})".format(table)
    schema = pd.read_sql_query(query,con)
    print("Schema of",table)
    display(schema)
    print("-"*100)
    print("\n")
```

## **Exercise**

### **Execute a failing query (i.e. one which gives an error) to retrieve all employees records whose salary is lower than the average salary.**

```python
try:  
  pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE
                      WHERE SALARY<AVG(SALARY)
                      
                      """, con)
except:
  print("intended Error")
```

### **Execute a working query using a sub-select to retrieve all employees records whose salary is lower than the average salary.**

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE
                      WHERE SALARY<(SELECT AVG(SALARY) FROM EMPLOYEE)
                      
                      """, con)
```

### **Execute a failing query (i.e. one which gives an error) to retrieve all employees records with EMP_ID, SALARY and maximum salary as MAX_SALARY in every row.**

```python
pd.read_sql_query("""
                      SELECT EMP_ID, SALARY,MAX(SALARY) AS MAX_SALARY FROM EMPLOYEE
                         
                      """, con)
```

### **Execute a Column Expression that retrieves all employees records with EMP_ID, SALARY and maximum salary as MAX_SALARY in every row.**

```python
pd.read_sql_query("""
                      SELECT EMP_ID, SALARY,
                      (SELECT MAX(SALARY) FROM EMPLOYEE) AS MAX_SALARY 
                      FROM EMPLOYEE
                      
                      
                      """, con)
```

# Hands-on Lab: Working with Multiple Tables

## Exercise 1: Accessing Multiple Tables with Sub-Queries

### Retrieve only the EMPLOYEES records that correspond to jobs in the JOBS table.

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE

                      """, con)
```

```python
pd.read_sql_query("""
                      SELECT * FROM JOBS

                      """, con)
```

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE
                      WHERE JOB_ID IN (SELECT JOB_IDENT FROM JOBS)

                      """, con)
```

### Retrieve only the list of employees whose JOB_TITLE is Jr. Designer.

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE
                      WHERE JOB_ID IN 
                            (SELECT JOB_IDENT FROM JOBS 
                            WHERE JOB_TITLE= 'Jr. Designer')

                      """, con)
```

### Retrieve JOB information and list of employees who earn more than $70,000.

```python
pd.read_sql_query("""
                      SELECT JOB_TITLE, MIN_SALARY,MAX_SALARY, JOB_IDENT FROM JOBS 
                      WHERE JOB_IDENT IN
                        (SELECT JOB_ID FROM EMPLOYEE WHERE SALARY>70000)

                      """, con)
```

### Retrieve JOB information and list of employees whose birth year is after 1976.

```python
pd.read_sql_query("""
                      SELECT JOB_TITLE, MIN_SALARY,MAX_SALARY, JOB_IDENT FROM JOBS 
                      WHERE JOB_IDENT IN
                        (SELECT JOB_ID FROM EMPLOYEE WHERE B_DATE>1976)

                      """, con)
```

### Retrieve JOB information and list of female employees whose birth year is after 1976.

```python
pd.read_sql_query("""
                      SELECT JOB_TITLE, MIN_SALARY,MAX_SALARY, JOB_IDENT FROM JOBS 
                      WHERE JOB_IDENT IN
                        (SELECT JOB_ID FROM EMPLOYEE WHERE B_DATE>1976 AND SEX= 'F')

                      """, con)
```

## Exercise 2: Accessing Multiple Tables with Implicit Joins


### Perform an implicit cartesian/cross join between EMPLOYEES and JOBS tables.

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE, JOBS
                      """, con)
```

```python
pd.read_sql_query("""
                      SELECT * FROM JOBS,EMPLOYEE
                      """, con)
```

### Retrieve only the EMPLOYEES records that correspond to jobs in the JOBS table.

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE, JOBS
                      WHERE JOBS.JOB_IDENT=EMPLOYEE.JOB_ID

                      """, con)
```

### Redo the previous query, using shorter aliases for table names.

```python
pd.read_sql_query("""
                      SELECT * FROM EMPLOYEE e, JOBS j
                      WHERE j.JOB_IDENT=e.JOB_ID

                      """, con)
```

### Redo the previous query, but retrieve only the Employee ID, Employee Name and Job Title.

```python
pd.read_sql_query("""
                      SELECT EMP_ID, F_NAME, L_NAME, JOB_TITLE FROM EMPLOYEE e, JOBS j
                      WHERE j.JOB_IDENT=e.JOB_ID

                      """, con)
```

### Redo the previous query, but specify the fully qualified column names with aliases in the SELECT clause.

```python
pd.read_sql_query("""
                      SELECT e.EMP_ID, e.F_NAME, e.L_NAME, j.JOB_TITLE FROM EMPLOYEE e, JOBS j
                      WHERE j.JOB_IDENT=e.JOB_ID

                      """, con)
```
